```python
from decimal import *

getcontext().prec = 102
squares = [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
```

```python
total = 0
for x in range(100):
    if x not in squares:
        root = Decimal(x).sqrt()
        print(root)
        parsed = str(root).replace('.', '')[:100]
        print(parsed)
        digits = [int(char) for char in parsed]
        print(sum(digits))
        total += sum(digits)
print(total)
```
