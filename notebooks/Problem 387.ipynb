```python
def gen_primes():
    """ Generate an infinite sequence of prime numbers """
    d, q = {}, 2
    while True:
        if q not in d:
            yield q
            d[q * q] = [q]
        else:
            for p in d[q]:
                d.setdefault(p + q, []).append(p)
            del d[q]
        q += 1
```

```python
def is_harshad(num):
    num_digits = [int(x) for x in str(num)]
    return not int(''.join(str(x) for x in num_digits)) % sum(num_digits)
```

```python
primer, max_prime = gen_primes(), 2012
prime = next(primer)
primes, targets = [prime], []
while prime < max_prime:
    
    prime = next(primer)
    primes.append(prime)
print(sum(targets))
```
